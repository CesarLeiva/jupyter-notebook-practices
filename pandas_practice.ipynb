```python
#Concatenar DaraFrames por filas
#Para esto deben tener el mismo índice de columnas

x = pd.DataFrame({'NOMBRE':['César','María','Alex'],
                  'EDAD':[21,22,16],
                  'CARRERA':['Ing Sis','Historia','Derecho']}).set_index('NOMBRE')
x2 =pd.DataFrame({'NOMBRE':['Juan','Pedro','Luis'],
                  'EDAD':[23,28,26],
                  'CARRERA':['Periodismo','Ing Civil','Historia']}).set_index('NOMBRE')
#El .set_index() ayuda a concatenar los DataFrame por medio de un punto específico

y = pd.concat([x, x2])
y
```

```python
#Combinación por columna
#Para esto deben tener el mismo índice de filas

m = pd.DataFrame({'MARCA':['Lenovo','Asus','Hp'],
                  'RAM':[16,20,8],
                  'PESO':[1.6,1.3,1.8]}).set_index('MARCA')

m2 = pd.DataFrame({'MARCA':['Lenovo','Asus','Hp'],
                  'REFERENCIA':['ABD-12','EAR4F','MR-3ER'],
                  'PROCESADOR':['Core i5','Ryzen 7','Ryzen 5']}).set_index('MARCA')

pd.concat([m,m2], axis=1)
```

```python
#Las convinaciones se hacen mediante una clave y mezclan varos dataframes
#sólo mezcla la información en común
pd.merge(m,m2,on='MARCA',how='outer')

#Con how='inner' se mezclan por columnas (este viene por defecto)
#con how='outer' se mezclan por filas
#con how='left' o how='right', se toma como referencia principal uno u otro DataFrame
```
